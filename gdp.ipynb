```python
import pandas as pd
from pathlib import Path
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
path="csv_data/states.csv"
state_ab=pd.read_csv(path,encoding="ISO-8859-1")
pd.set_option("display.max_rows",500)
pd.set_option("display.max_columns",500)
```

```python
path = "csv_data/GDP_States_2017_2022.csv"
gdp_data=pd.read_csv(path,encoding="ISO-8859-1")
pd.set_option("display.max_rows",500)
pd.set_option("display.max_columns",500)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
gdp_data.head(5)
```

```python
new=(gdp_data["LineCode"] == 1.00)
state_gdp = gdp_data.loc[new]
col=["GeoName","Unit","2017","2018","2019","2020","2021","2022"]
state_gdp=state_gdp[col]
state_gdp=state_gdp.rename(columns={"GeoName":"State"})
state_gdp.set_index("State").head()
```

```python
state_ab.head()
```

```python
merge_df1 = pd.merge(left= state_gdp, right=state_ab,on="State",how="left")
merge_df1
```

```python
merge_df1= merge_df1.dropna(how="any")
merge_df1 =merge_df1[["State","Abbreviation","Unit","2022"]]
merge_df1.sort_values(by="2022",ascending=False)
merge_df1.head()
```

```python
path="production.csv"
production=pd.read_csv(path,encoding="ISO-8859-1")
pd.set_option("display.max_rows",500)
pd.set_option("display.max_columns",500)
```

```python

production=production.rename(columns={"State":"Abbreviation"})
production.head()
```

```python
merge_df2 = pd.merge(left= production, right=merge_df1,on="Abbreviation",how="left")
```

```python
merge_df2.dropna(how='any')
```

```python
merge_df2=merge_df2[["State","Production_WSTE","2022"]]
#merge_df2["Production_WSTE"]=merge_df2["Production_WSTE"]/1000
merge_df2["Production_WSTE"]=merge_df2["Production_WSTE"].astype(int)
#merge_df2["2022"]=merge_df2["2022"].astype(int)
merge_df2
```
